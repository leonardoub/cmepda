#ANOVA SCIPY

```
# load packages
import scipy.stats as stats

dict_feat_anova = {}
ANOVA_test_features = []


for column in data.columns:

  #feat_A = public_data[column][public_data.Histology == 'adenocarcinoma']
  feat_GBM = data[column][data_1.Histology == 'GBM']
  feat_LGG = data[column][data_1.Histology == 'LGG']

  fvalue, pvalue = stats.f_oneway(feat_GBM, feat_LGG)

  if pvalue <= 0.05:
    print(fvalue, pvalue, column)
    ANOVA_test_features.append(column)

    dict_feat_anova.update({column : pvalue})

  #print(fvalue, pvalue, column)
```

```
df_feat_anova = pd.DataFrame.from_dict(dict_feat_anova, orient='index', columns=['p_value'])
df_feat_anova_sorted = df_feat_anova.sort_values(by='p_value', ascending=False)
```


#T-test

```
from scipy.stats import ttest_ind

# load packages
import scipy.stats as stats

dict_feat_T_test = {}
T_test_features = []

for column in data.columns:

  #feat_A = public_data[column][public_data.Histology == 'adenocarcinoma']
  feat_GBM = data[column][data_1.Histology == 'GBM']
  feat_LGG = data[column][data_1.Histology == 'LGG']

  fvalue, pvalue = stats.ttest_ind(feat_GBM, feat_LGG, equal_var=True)

  if pvalue <= 0.05:
    print(fvalue, pvalue, column)
    T_test_features.append(column)

    dict_feat_T_test.update({column : pvalue})

  #print(fvalue, pvalue, column)
```

```
df_feat_T_test = pd.DataFrame.from_dict(dict_feat_T_test, orient='index', columns=['p_value'])
df_feat_T_test_sorted = df_feat_T_test.sort_values(by='p_value', ascending=False)
```

#Welch’s t-test, which does not assume equal population variance


```
from scipy.stats import ttest_ind

# load packages
import scipy.stats as stats

dict_feat_Welch_test = {}
Welch_test_features = []


for column in data.columns:

  #feat_A = public_data[column][public_data.Histology == 'adenocarcinoma']
  feat_GBM = data[column][data_1.Histology == 'GBM']
  feat_LGG = data[column][data_1.Histology == 'LGG']

  fvalue, pvalue = stats.ttest_ind(feat_GBM, feat_LGG, equal_var=False)

  if pvalue <= 0.05:
    print(fvalue, pvalue, column)
    Welch_test_features.append(column)

    dict_feat_Welch_test.update({column : pvalue})

  #print(fvalue, pvalue, column)
```

```
df_feat_Welch_test = pd.DataFrame.from_dict(dict_feat_Welch_test, orient='index', columns=['p_value'])
df_feat_Welch_test_sorted = df_feat_Welch_test.sort_values(by='p_value', ascending=False)
```


#Test di Mann-Whitney

The Mann-Whitney U test for comparing independent data samples: the nonparametric version of the Student t-test.
Nonparametric statistics are those methods that do not assume a specific distribution to the data.


The null hypothesis of these tests is often the assumption that both samples were drawn from a population with the same distribution, and therefore the same population parameters, such as mean or median.

```
# load packages
import scipy.stats as stats

dict_feat_M_W_test = {}
M_W_test_features = []


for column in data.columns:

  #feat_A = public_data[column][public_data.Histology == 'adenocarcinoma']
  feat_GBM = data[column][data_1.Histology == 'GBM']
  feat_LGG = data[column][data_1.Histology == 'LGG']

  fvalue, pvalue = stats.mannwhitneyu(feat_GBM, feat_LGG)

  if pvalue <= 0.05:
    print(fvalue, pvalue, column)
    M_W_test_features.append(column)

    dict_feat_M_W_test.update({column : pvalue})

  #print(fvalue, pvalue, column)
```

```
df_feat_M_W_test = pd.DataFrame.from_dict(dict_feat_M_W_test, orient='index', columns=['p_value'])
df_feat_M_W_test_sorted = df_feat_M_W_test.sort_values(by='p_value', ascending=False)
```

```
from sklearn.feature_selection import SelectKBest, SelectPercentile
from sklearn.feature_selection import f_classif
```

```
# Create an SelectKBest object to select features with two best ANOVA F-Values
fvalue_selector = SelectKBest(f_classif, k=8)
fvalue_selector_percentile = SelectPercentile(f_classif, percentile=10)
```
